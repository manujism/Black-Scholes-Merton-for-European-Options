```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Black-Scholes Pricing ---
def black_scholes(s, k, r, t, sigma, option_type='Call'):
    d1 = (np.log(s / k) + (r + (sigma ** 2) / 2) * t) / (sigma * np.sqrt(t))
    d2 = d1 - sigma * np.sqrt(t)

    if option_type == 'Call':
        price = s * norm.cdf(d1) - k * np.exp(-r * t) * norm.cdf(d2)
    else:
        price = k * np.exp(-r * t) * norm.cdf(-d2) - s * norm.cdf(-d1)

    return d1, d2, price

# --- Greeks Calculation ---
def calculate_greeks(d1, d2, s, k, r, t, sigma, option_type='Call'):
    pdf_d1 = norm.pdf(d1)
    gamma = pdf_d1 / (s * sigma * np.sqrt(t))
    vega = s * np.sqrt(t) * pdf_d1

    if option_type == 'Call':
        delta = norm.cdf(d1)
        theta = (- (s * pdf_d1 * sigma) / (2 * np.sqrt(t))
                 - r * k * np.exp(-r * t) * norm.cdf(d2))
        rho = k * t * np.exp(-r * t) * norm.cdf(d2)
    else:
        delta = norm.cdf(d1) - 1
        theta = (- (s * pdf_d1 * sigma) / (2 * np.sqrt(t))
                 + r * k * np.exp(-r * t) * norm.cdf(-d2))
        rho = -k * t * np.exp(-r * t) * norm.cdf(-d2)

    return delta, gamma, vega, theta, rho

# --- Parameters ---
np.random.seed(0)
k = np.random.uniform(50, 150)       # Strike price
t = np.random.uniform(0.1, 2)        # Time to maturity
r = np.random.uniform(0.01, 0.05)    # Risk-free rate
sigma = np.random.uniform(0.1, 0.5)  # Volatility
s_values = np.linspace(0.1, 200, 200)  # Stock prices

# --- Containers for Greeks (Call and Put) ---
call_greeks = {'delta': [], 'gamma': [], 'vega': [], 'theta': [], 'rho': []}
put_greeks  = {'delta': [], 'gamma': [], 'vega': [], 'theta': [], 'rho': []}

# --- Compute Greeks for all s_values ---
for s in s_values:
    for option_type, greek_dict in [('Call', call_greeks), ('Put', put_greeks)]:
        d1, d2, _ = black_scholes(s, k, r, t, sigma, option_type)
        delta, gamma, vega, theta, rho = calculate_greeks(d1, d2, s, k, r, t, sigma, option_type)

        greek_dict['delta'].append(delta)
        greek_dict['gamma'].append(gamma)
        greek_dict['vega'].append(vega)
        greek_dict['theta'].append(theta)
        greek_dict['rho'].append(rho)

# --- Plotting ---
plt.figure(figsize=(16, 12))

# Helper to plot both call and put on same subplot
def plot_greek(index, greek_name, ylabel, color1, color2):
    plt.subplot(3, 2, index)
    plt.plot(s_values, call_greeks[greek_name], label='Call', color=color1)
    plt.plot(s_values, put_greeks[greek_name], label='Put', color=color2, linestyle='--')
    plt.title(f'{greek_name.capitalize()} vs Stock Price')
    plt.xlabel('Stock Price')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend()

# Plot all Greeks
plot_greek(1, 'delta', 'Delta', 'blue', 'red')
plot_greek(2, 'gamma', 'Gamma', 'green', 'orange')
plot_greek(3, 'vega', 'Vega', 'purple', 'brown')
plot_greek(4, 'theta', 'Theta', 'teal', 'magenta')
plot_greek(5, 'rho', 'Rho', 'darkblue', 'darkred')

plt.tight_layout()
plt.show()

call_prices = []
put_prices = []

for s in s_values:
    _, _, call_price = black_scholes(s, k, r, t, sigma, 'Call')
    _, _, put_price  = black_scholes(s, k, r, t, sigma, 'Put')
    call_prices.append(call_price)
    put_prices.append(put_price)

# --- Plotting Prices ---
plt.figure(figsize=(10, 6))
plt.plot(s_values, call_prices, label='Call Option Price', color='blue')
plt.plot(s_values, put_prices, label='Put Option Price', color='red', linestyle='--')
plt.title('Option Prices vs Stock Price')
plt.xlabel('Stock Price')
plt.ylabel('Option Price')
plt.grid(True)
plt.legend()
plt.show()
```
